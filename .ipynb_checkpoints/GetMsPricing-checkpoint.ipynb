```python
try:
    from bs4 import BeautifulSoup
except ImportError:
    from BeautifulSoup import BeautifulSoup
import requests
import json
import pandas as pd
```

```python
url = "https://azure.microsoft.com/fr-fr/pricing/details/virtual-machines/linux/"
```

```python
page = requests.get(url)
print(page.content)
```

```python
soup = BeautifulSoup(page.content)
```

```python
tables = soup.findAll("table", {"class": "data-table__table data-table__table--default"})
#tables = soup.findAll("tbody")
```

```python
print(tables)
```

```python
output = list()
for table in tables:
    trs = table.findAll("tr")
    for tr in trs:
        row= list()
        tds = tr.findAll("td")
        for td in tds:
            try:
                row.append(json.loads(td.find('span')['data-amount'])['regional']['europe-west'])
            except:
                row.append(td.text.strip())
        if len(row) > 1:
            output.append(row)
```

```python
print(output)
```

```python
df = pd.DataFrame(output)
```

```python
df
```

```python
df.drop([0], inplace=True, axis=1)
```

```python
df = df.loc[:, :8]
```

```python
df.columns = ["instance", "vCPU", "RAM", "Storage", "hourly", "1y", "2y", "3y"]
```

```python
df = df.replace({'– –\nVide': 'na'}, regex=True)
```

```python
df = df.replace({'Gio': ''}, regex=True)
```
